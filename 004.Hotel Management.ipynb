```python
import datetime
ch=0
while ch!=6:
    print('\n')
    print(" 1. DATE OF CHECK IN")
    print("2. DATE OF CHECK OUT")
    print("3. COST OF ROOM")
    print("4. OTHER EXPENSES")
    print('5. GENERATE BILL')
    print('6. TERMINATE')
    print("\n")
    ch=int(input("ENTER YOUR CHOICE:"))
    if ch==1:
        print('ENTER CHECK-IN DATE IN YYYY-MM-DD FORMAT:')
        d=int(input('ENTER DAY:'))
        m=int(input('ENTER MONTH:'))
        year=int(input('ENTER YEAR:'))
        print('\n')
        ind=datetime.date(year,m,d)
    elif ch==2:
       fd= datetime.date.today()
       print("CHECK-OUT DATE IS",fd)
    elif ch==3:
        no_day= fd-ind
        print("NUMBER OF DAYS:",no_day)
        nd=int(input('ENTER COST OF ROOM:'))
        e=no_day.days
        c= nd*e
        print("BILL FOR THE ROOM:",c)
    elif ch==4:
        o= int(input("ENTER OTHER EXPENSES:"))
    elif ch==5:
        b=c+o
        print("\n")
        print('FINAL BILL IS \n')
        print('COST OF ROOM:',c)
        print("OTHER EXPENSES:",o)
        print('TOTAL AMOUNT:',b)
    elif ch==6:
        print("PROGRAMME TERMINATED")
        break
    else:
        print("ERROR! ENTER A VALID CHOICE")
```
